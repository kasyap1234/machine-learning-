```python
X=data.drop('Heart Disease',axis=1)
```

```python
encoded_data=pd.get_dummies(data,columns=['Heart Disease'])
encoded_data.head()
```

```python
Y=encoded_data['Heart Disease_Presence']
```

***label encoding needs to be performed to remove the string type categorical feature*********

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.1,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression 
lor=LogisticRegression(solver='liblinear')
```

```python
xtrain
```

```python
ytrain
```

```python
lor.fit(xtrain,ytrain)
```

```python
ypred=lor.predict(xtest)
```

```python
lor.score(xtest,ytest)
```
